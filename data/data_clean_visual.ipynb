## Notebook per preprocessar i visualtzar el datasaet de yfinance

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Llegir l'axiu amb al dataset
df = pd.read_parquet('SP500.parquet')

# Ajuntar Columnes per practicitat ja que es mulitcolumna
df.columns = ['_'.join(col) for col in df.columns]

print(df.size)
```

## Neteja i Enriquiment de les Dades

```python
print(df)
```

## Visualització del Preu mitjà pensual per any de tancament de l'S&P500

```python
anual_average_value = df.groupby(df.index.year)['Close_^GSPC'].mean()


plt.bar(anual_average_value.index, anual_average_value.values)
plt.xlabel('Any')
plt.ylabel('Preu mitjà mensual per any ($)')
plt.title('Preu mitjà mensual per any de tancament de S&P500')
plt.xlim(anual_average_value.index.min() - 0.5, anual_average_value.index.max() + 0.5)
plt.show()
```
